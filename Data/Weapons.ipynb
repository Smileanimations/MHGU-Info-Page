# MHGU Weapons
We importeren **pandas** en **matplotlib** voor onze visualisatie.

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
```

## Weapons.csv
We importeren **"Weapons.csv"** in onze notebook.

```python
df = pd.read_csv('Weapons.csv')
df.head()
```

## MHGU Weapon popularity
We maken de taartdiagram, waarbij de percentages automatisch worden berekend.

```python
# Extract the data for the pie chart
names = df['WEAPON_CLASS']
size = df['WEAPON_POPULARITY'] 

# Maak een figuur met een zwarte achtergrond
fig, ax = plt.subplots(figsize=(10, 8), facecolor='black')
ax.set_facecolor('black')

# Maak de donut-plot
wedges, texts, autotexts = ax.pie(
    size, 
    wedgeprops={'linewidth': 2, 'edgecolor': 'white'}, 
    autopct='%1.1f%%',
    pctdistance=0.85,
    textprops={'color': 'black'},
)

# Voeg een cirkel toe in het midden om het donut-effect te creëren
my_circle = plt.Circle((0, 0), 0.70, color='black', fc='black')
ax.add_artist(my_circle)

# Pas de tekstkleur aan voor de labels
for text in texts:
    text.set_color('white')

# Voeg een legenda toe
plt.legend(
    wedges,
    names, 
    title="WEAPON CLASS", 
    loc="center left",
    bbox_to_anchor=(1, 0.5),
    facecolor='white',
    title_fontsize='14',
    fontsize='11',
    labelcolor='black'
)

# Toon de plot
plt.tight_layout()
plt.show()
```

## Wat je kunt opmerken
* **Populairste Wapen:** In onze dataset is de **'Greatsword'** het populairste wapen met een populariteit van **13,867.** Dit wordt weergegeven als het grootste segment in de taartdiagram.

* **Minder Populaire Wapens:** Wapens zoals **'Gunlance'** en **'Hunting Horn'** hebben een veel lagere populariteit (respectievelijk **1,362** en **2,178**), en hun segmenten zullen aanzienlijk kleiner zijn in de grafiek.

## Conclusie
Een taartdiagram biedt een duidelijke en visuele manier om te zien hoe de populariteit van verschillende MHGU wapens zich tot elkaar verhoudt. Het is een handig hulpmiddel voor het analyseren van gegevens en het nemen van beslissingen op basis van de populariteit van wapens. Door visueel inzicht te krijgen in deze verhoudingen, kan je beter begrijpen welke wapens het meest en het minst populair zijn binnen de gegeven dataset.
